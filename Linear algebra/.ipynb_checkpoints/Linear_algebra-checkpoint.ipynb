## Inner product and projection definition for real and complex numbers 

### 1. Inner product and projection defintion for real numbers

### 1.1 Inner product definition
The inner product (or dot product) for real vectors in a real vector space is defined as follows:

Given two vectors $ \mathbf{u} $ and $ \mathbf{v} $ in a real vector space $ \mathbb{R}^n $, the inner product $ \langle \mathbf{u}, \mathbf{v} \rangle $ is:

$$
\langle \mathbf{u}, \mathbf{v} \rangle = \sum_{i=1}^n u_i v_i
$$

where $ u_i $ and $ v_i $ are the $ i $ -th components of the vectors $ \mathbf{u} $ and $ \mathbf{v} $, respectively.

#### 1.1.1 Properties

The inner product of two real vectors satisfies the following properties:

1. **Commutativity**:
   $$
   \langle \mathbf{u}, \mathbf{v} \rangle = \langle \mathbf{v}, \mathbf{u} \rangle
   $$

2. **Linearity**:
   $$
   \langle a\mathbf{u} + b\mathbf{w}, \mathbf{v} \rangle = a\langle \mathbf{u}, \mathbf{v} \rangle + b\langle \mathbf{w}, \mathbf{v} \rangle
   $$
   for all vectors $ \mathbf{u}, \mathbf{w}, \mathbf{v} \in \mathbb{R}^n $ and scalars $ a, b \in \mathbb{R} $.

3. **Positive-definiteness**:
   $$
   \langle \mathbf{u}, \mathbf{u} \rangle \geq 0
   $$
   with equality if and only if $ \mathbf{u} = 0 $.

#### 1.1.2 Example

For vectors $ \mathbf{u} = (u_1, u_2, \ldots, u_n) $ and $ \mathbf{v} = (v_1, v_2, \ldots, v_n) $ in $ \mathbb{R}^n \$, the inner product is calculated as follows:

$$
\langle \mathbf{u}, \mathbf{v} \rangle = u_1 v_1 + u_2 v_2 + \cdots + u_n v_n
$$

### 1.2 Projection definition
Given two vectors $ \mathbf{u} $ and $ \mathbf{v} $ in a real vector space $ \mathbb{R}^n $, the projection of $ \mathbf{u} $ onto $ \mathbf{v} $, denoted as $ \text{proj}_{\mathbf{v}} \mathbf{u} $, is defined as:

$$
\text{proj}_{\mathbf{v}} \mathbf{u} = \frac{\langle \mathbf{u}, \mathbf{v} \rangle}{\langle \mathbf{v}, \mathbf{v} \rangle} \mathbf{v}
$$

where $ \langle \mathbf{u}, \mathbf{v} \rangle $ is the inner product (dot product) of $ \mathbf{u} $ and $\mathbf{v}$, defined as:

$$
\langle \mathbf{u}, \mathbf{v} \rangle = \sum_{i=1}^n u_i v_i
$$

#### 1.2.1 Example

For vectors $ \mathbf{u} = (u_1, u_2, \ldots, u_n) $ and $ \mathbf{v} = (v_1, v_2, \ldots, v_n) $ in $ \mathbb{R}^n $:

1. Compute the inner product $ \langle \mathbf{u}, \mathbf{v} \rangle $:

$$
\langle \mathbf{u}, \mathbf{v} \rangle = u_1 v_1 + u_2 v_2 + \cdots + u_n v_n
$$

2. Compute the inner product $\langle \mathbf{v}, \mathbf{v} \rangle $:

$$
\langle \mathbf{v}, \mathbf{v} \rangle = v_1^2 + v_2^2 + \cdots + v_n^2
$$

3. Compute the projection:

$$
\text{proj}_{\mathbf{v}} \mathbf{u} = \frac{u_1 v_1 + u_2 v_2 + \cdots + u_n v_n}{v_1^2 + v_2^2 + \cdots + v_n^2} \mathbf{v}
$$

4. Multiply by $ \mathbf{v} $:

$$
\text{proj}_{\mathbf{v}} \mathbf{u} = \left( \frac{u_1 v_1 + u_2 v_2 + \cdots + u_n v_n}{v_1^2 + v_2^2 + \cdots + v_n^2} \right) (v_1, v_2, \ldots, v_n)
$$

### 2. Inner product and projection defintion for Complex numbers

### 2.1 Inner product definition for complex numbers

The inner product (or Hermitian inner product) for complex vectors in a complex vector space is defined as follows:

Given two vectors $ \mathbf{u} $ and $ \mathbf{v} $ in a complex vector space $ \mathbb{C}^n $, the inner product $ \langle \mathbf{u}, \mathbf{v} \rangle $ is:

$$
\langle \mathbf{u}, \mathbf{v} \rangle = \sum_{i=1}^n u_i \overline{v_i}
$$

where $ \overline{v_i} $ denotes the complex conjugate of the $ i $-th component of $ \mathbf{v} $.

#### 2.1.1 Properties

The inner product of two complex vectors satisfies the following properties:

1. **Conjugate Symmetry**:
   $$
   \langle \mathbf{u}, \mathbf{v} \rangle = \overline{\langle \mathbf{v}, \mathbf{u} \rangle}
   $$

2. **Linearity in the First Argument**:
   $$
   \langle a\mathbf{u} + b\mathbf{w}, \mathbf{v} \rangle = a\langle \mathbf{u}, \mathbf{v} \rangle + b\langle \mathbf{w}, \mathbf{v} \rangle
   $$
   for all vectors $ \mathbf{u}, \mathbf{w}, \mathbf{v} \in \mathbb{C}^n $ and scalars $ a, b \in \mathbb{C} $.

3. **Positive-Definiteness**:
   $$
   \langle \mathbf{u}, \mathbf{u} \rangle \geq 0
   $$
   with equality if and only if $ \mathbf{u} = \mathbf{0} $.

#### 2.1.2 Example

For vectors $ \mathbf{u} = (u_1, u_2, \ldots, u_n) $ and $ \mathbf{v} = (v_1, v_2, \ldots, v_n) $ in $ \mathbb{C}^n $, the inner product is calculated as follows:

$$
\langle \mathbf{u}, \mathbf{v} \rangle = u_1 \overline{v_1} + u_2 \overline{v_2} + \cdots + u_n \overline{v_n}
$$




### 2.2 Projection definition for complex numbers

Given two vectors $ \mathbf{u} $ and $ \mathbf{v} $ in a complex vector space $ \mathbb{C}^n $, the projection of $ \mathbf{u} $ onto $ \mathbf{v} $, denoted as $ \text{proj}_{\mathbf{v}} \mathbf{u} $, is defined as:

$$
\text{proj}_{\mathbf{v}} \mathbf{u} = \frac{\langle \mathbf{u}, \mathbf{v} \rangle}{\langle \mathbf{v}, \mathbf{v} \rangle} \mathbf{v}
$$

where $ \langle \mathbf{u}, \mathbf{v} \rangle $ is the Hermitian inner product of $ \mathbf{u} $ and $ \mathbf{v} $, defined as:

$$
\langle \mathbf{u}, \mathbf{v} \rangle = \sum_{i=1}^n u_i \overline{v_i}
$$

#### 2.2.1 Example

For vectors $ \mathbf{u} = (u_1, u_2, \ldots, u_n) $ and $ \mathbf{v} = (v_1, v_2, \ldots, v_n) $ in $ \mathbb{C}^n $:

1. Compute the Hermitian inner product $ \langle \mathbf{u}, \mathbf{v} \rangle $:

$$
\langle \mathbf{u}, \mathbf{v} \rangle = u_1 \overline{v_1} + u_2 \overline{v_2} + \cdots + u_n \overline{v_n}
$$

2. Compute the Hermitian inner product $ \langle \mathbf{v}, \mathbf{v} \rangle $:

$$
\langle \mathbf{v}, \mathbf{v} \rangle = v_1 \overline{v_1} + v_2 \overline{v_2} + \cdots + v_n \overline{v_n}
$$

3. Compute the projection:

$$
\text{proj}_{\mathbf{v}} \mathbf{u} = \frac{u_1 \overline{v_1} + u_2 \overline{v_2} + \cdots + u_n \overline{v_n}}{v_1 \overline{v_1} + v_2 \overline{v_2} + \cdots + v_n \overline{v_n}} \mathbf{v}
$$

4. Multiply by $ \mathbf{v} $:

$$
\text{proj}_{\mathbf{v}} \mathbf{u} = \left( \frac{u_1 \overline{v_1} + u_2 \overline{v_2} + \cdots + u_n \overline{v_n}}{v_1 \overline{v_1} + v_2 \overline{v_2} + \cdots + v_n \overline{v_n}} \right) (v_1, v_2, \ldots, v_n)
$$

```python
###################   Given #################################
#Original function
L = np.pi #it is from range -L to L
dx = 0.0001
x_range = L*(np.arange(-1,1,dx)) # domain [-L,L]
A0_given = 2 #intial value

harmonic_cos = 1  # original harmonics used for constructing the target function
harmonic_sin = 3 # original harmonics used for constructing the target function

print("The harmonic number for cosine is:",harmonic_cos)
print("The harmonic number for sine is:",harmonic_sin,'\n')
###########################################################################

omega_cos = harmonic_cos*2*np.pi/(2*L) #angular frequency (note the period is l--L = 2L
omega_sin = harmonic_sin*2*np.pi/(2*L) #angular frequency

print("The orginal omega(angular frequency for cos is:",omega_cos)
print("The orginal omega(angular frequency for sin is:",omega_sin,'\n')

Amplitude_sin =5
Amplitude_cosine = 6

original_func = A0_given + Amplitude_sin*np.sin(omega_sin*x_range) + Amplitude_cosine*np.cos(omega_cos*x_range)
#Plot the original function
plt.plot(original_func)
plt.xlabel("Time(s)")
plt.ylabel("Amplitude")
```

### 3. Fourier series
The basis vectors of the Fourier transform, which are complex exponentials of the form 
$e^{i 2 \pi k x}$, are unique and fundamental in several ways:
### &emsp;&emsp;&emsp; 1. Orthogonality

The basis vectors $ e^{i 2 \pi k x} $ are orthogonal over the interval $[0, 1]\):

$$
\int_{0}^{1} e^{i 2 \pi k x} \overline{e^{i 2 \pi m x}} \, dx = 0 \quad \text{for} \quad k \neq m
$$

### &emsp;&emsp;&emsp; 2. Completeness

Any function $ f(x) $ in the space of square-integrable functions $ L^2([0, 1]) $ can be expressed as:

$$
f(x) = \sum_{k=-\infty}^{\infty} c_k e^{i 2 \pi k x}
$$

For the continuous Fourier transform:

$$
f(x) = \int_{-\infty}^{\infty} \hat{f}(\xi) e^{i 2 \pi \xi x} \, d\xi
$$

### &emsp;&emsp;&emsp; 3. Frequency Interpretation

The basis vectors correspond to sinusoidal functions with frequencies \( k \) (for the Fourier series) or \( \xi \) (for the Fourier transform).

### &emsp;&emsp;&emsp; 4. Invariance under Translation and Scaling

- Translation in the time domain corresponds to a phase shift in the frequency domain.
- Scaling in the time domain corresponds to scaling in the frequency domain.

### &emsp;&emsp;&emsp; 5. Efficiency in Computation

The Fast Fourier Transform (FFT) algorithm leverages the structure of the Fourier basis functions to efficiently compute the Discrete Fourier Transform (DFT) with complexity \( O(n \log n) \).

#### 3.1.3 Fourier coefficients computation
 Projecting the **real** function f(x) onto different base vectors to find the component of the function along each base vector which are orthogonal
 
##### &emsp; Projecting a onto base vector b using inner product
&emsp;&emsp;&emsp;&emsp; $ \text{proj}_{\mathbf{b}} \mathbf{a} = \frac{\langle \mathbf{a}, \mathbf{b} \rangle}{\langle \mathbf{b}, \mathbf{b} \rangle} \mathbf{b} $

##### &emsp; The inner product of two **real** functions \(a(x)\) and \(b(x)\) over the **real number interval** \([a, b]\) is given by:
&emsp;&emsp;&emsp;&emsp; $ \langle a, b \rangle = \int_a^b a(x) b(x) \, dx $


##### &emsp; If domain x = [-L, L]
&emsp;&emsp;&emsp;&emsp; $A_{0}=\frac{\langle \mathbf{f(x)}, \mathbf{1} \rangle}{\langle \mathbf{1}, \mathbf{1} \rangle} = \frac{\langle f, 1 \rangle}{\langle 1, 1 \rangle}= \frac{\int_{-L}^L f(x) \, dx}{L - -L}  = \frac{1}{2L}\int_{L}^{-L}f(x) dx$ 


<blockquote>
    <blockquote>
"`When we project a function \( f(x) \) onto \( g(x) = 1 \), we isolate the component of the function over an interval.This is analogous to projecting onto a constant baseline."`

Mathematically, this projection is given by:

&emsp;&emsp;&emsp;&emsp;$  \text{proj}_{1} f = \frac{\langle f, 1 \rangle}{\langle 1, 1 \rangle}  = \frac{\int_{-L}^{L} f(x) \, dx}{L - L} = \frac{1}{2L}\int_{L}^{-L}f(x) dx   $

This expression represents the  **average** value of f(x) over the interval [a, b]. The result of this projection is a constant function, which can be thought of as "flattening" f(x) onto the x-axis where the **height** of this flattened function is the **average value** of f(x).
</blockquote>
</blockquote>

&emsp;&emsp;&emsp;&emsp; $A_{n}=\frac{1}{L}\int_{L}^{-L}f(x)\cos(\frac{\pi kx}{L}) dx$ 
<blockquote>
    <blockquote>
        "When we project a function \( f(x) \) onto \( g(x) = \cos(x) \), we isolate the component of the function over an interval on the cosine function. "

Mathematically, this projection is given by:

&emsp;&emsp;&emsp;&emsp;$  \text{proj}_{\cos(x)} f = \frac{\langle f, \cos(x) \rangle}{\langle \cos(x), \cos(x) \rangle} = \frac{\int_{-L}^{L} f(x) \cos(x) \, dx}{\int_a^b \cos^2(x) \, dx}   $

The result of this projection is a function represented by the cosine basis, which captures the periodicity of \( f(x) \) and its correlation with the cosine function.
</blockquote>
</blockquote>

&emsp;&emsp;&emsp;&emsp; $B_{n}=\frac{1}{L}\int_{L}^{-L}f(x)\cos(\frac{\pi kx}{L}) dx$ 

##### &emsp; If domain x = [-L,L] formula will be
&emsp;&emsp;&emsp;&emsp; $A_{0}=\frac{1}{2L}\int_{L}^{-L}f(x) dx$ 


&emsp;&emsp;&emsp;&emsp; $A_{n}=\frac{1}{L}\int_{L}^{-L}f(x)\cos(\frac{2 \pi kx}{L}) dx$ 


&emsp;&emsp;&emsp;&emsp; $B_{n}=\frac{1}{L}\int_{L}^{-L}f(x)\cos(\frac{2 \pi kx}{L}) dx$ 


##### &emsp; If domain x = [0,L] formula will be

### 3.1 Fourier function using exponential form
#### Key idea originates from euler formula that states:
&emsp; $ e^{ix} = \cos(x) + i\sin(x) $

Euler's formula provides a powerful connection between the complex exponential function and the trigonometric functions cosine and sine. The formulas are:

&emsp;$ e^{i\theta} = \cos(\theta) + i\sin(\theta) $

&emsp;$e^{-i\theta} = \cos(\theta) - i\sin(\theta) $

By adding and subtracting these equations, we can derive the expressions for \(\cos(\theta)\) and \(\sin(\theta)\) in terms of complex exponentials.

#### Derivation of $cos(\theta)$

Add the two equations:

&emsp;$ e^{i\theta} + e^{-i\theta} = (\cos(\theta) + i\sin(\theta)) + (\cos(\theta) - i\sin(\theta)) $

The imaginary parts cancel out, leaving:

&emsp;$ e^{i\theta} + e^{-i\theta} = 2\cos(\theta) $

Therefore:

&emsp;$ \cos(\theta) = \frac{e^{i\theta} + e^{-i\theta}}{2} = RE(e^{i\theta}) $

#### Derivation of $sin(\theta)$

Subtract the second equation from the first:

&emsp;$ e^{i\theta} - e^{-i\theta} = (\cos(\theta) + i\sin(\theta)) - (\cos(\theta) - i\sin(\theta)) $

The real parts cancel out, leaving:

&emsp;$ e^{i\theta} - e^{-i\theta} = 2i\sin(\theta) $

Therefore:

&emsp;$ \sin(\theta) = \frac{\large e^{i\theta} - e^{-i\theta}}{2i} = IM(e^{i\theta}) $

#### 3.2.1 Fourier function computation
#### Derivation of the Exponential Form of the Fourier Series using the Sine and Cosine formula

We start with the standard trigonometric form of the Fourier series for a periodic function \( f(t) \) with period \( T \):
##### &emsp; If domain x = [-L, L]
&emsp;&emsp;$f(x) = A_0 + \sum_{k=1}^{k=K} \left[ A_n \cos\left(\frac{\pi k x}{L}\right) + B_n \sin\left(\frac{\pi k x}{L}\right) \right]$

##### &emsp; If domain x = [0, L]
&emsp;&emsp;$f(x) = A_0 + \sum_{k=1}^{k=K} \left[ A_n \cos\left(\frac{2\pi k x}{L}\right) + B_n \sin\left(\frac{2\pi k x}{L}\right) \right]$

Inorder to derive a general formula let the expression within cosine and sine be n \
&emsp;$n = \frac{\pi k x}{L} $
&emsp;$n = \frac{2 \pi k x}{L} $

### Using Euler's Formulas

Euler's formulas relate the exponential function to sine and cosine:

### Express Cosine and Sine

Using these, we can express \(\cos(\theta)\) and \(\sin(\theta)\) in terms of exponentials:

$\cos\left(n\right) = \large\frac{e^{in} + e^{-in}}{2}$

$\sin\left(n\right) = \large\frac{e^{in} - e^{-in}}{2i}$

### Substitute into the Fourier Series

Substitute these into the trigonometric form of the Fourier series:

$$
f(x) = a_0 + \sum_{k=1}^{k=K} \left[ a_n (\frac{e^{in} + e^{-in}}{2}) + b_n (\frac{e^{in} - e^{-in}}{2i}) \right]
$$

### Combine and Simplify

Factor the sums involving exponentials:

$$
f(x) = a_0 + \sum_{k=1}^{k=K} \left[ \frac{a_n}{2} e^{in} + \frac{a_n}{2} e^{-in} + \frac{b_n}{2i} e^{in} - \frac{b_n}{2i} e^{-in} \right]
$$

$$
f(x) = a_0 + \sum_{k=1}^{k=K} \left[ \frac{a_n}{2} e^{in} + \frac{a_n}{2} e^{-in} + \frac{-i b_n}{2} e^{in} - \frac{-i b_n}{2} e^{-in} \right]
$$

$$
f(x) = a_0 + \sum_{k=1}^{k=K} \left[ \left( \frac{a_n - i b_n}{2}\right) e^{in} + \left( \frac{a_n + i b_n}{2} \right) e^{-in} \right]
$$

Instead of having to basis $e^{in}$ and $e^{-in}$ where $n = \frac{\pi k x}{L} $ or $ n = \frac{2 \pi k x}{L}$ change the k domain to be [-k,k] instead of [1,k] so that both bases are included within one formula and a_0 will also be represented when k=0


$$
f(x) = \sum_{k=-k}^{k=K} C_k e^{in} 
$$

Where \
&emsp;$C_k = \large \left( \frac{a_n - i b_n}{2}\right)$ &  $\large e^{in} = \frac{\pi |k| x}{L}$ or $ \large e^{in} = -\frac{2 \pi |k| x}{L}$ when $\large k>=0$ \
&emsp;$C_k = \large \left( \frac{a_n + i b_n}{2}\right)$ & $ \large e^{in} = \frac{- \pi |k| x}{L}$ or $ \large e^{in} = -\frac{2\pi |k| x}{L}$ when $\large k<0$

Note for real signals the imaginary part will be zero
$$
    f(x) = \sum_{k=-k}^{k=K} C_k e^{in} 
$$

$$
    f(x) = \sum_{k=0}^{k=K}  \left( \frac{a_n - i b_n}{2}\right) e^{in} + \sum_{k=0}^{k=K} \left( \frac{a_n + i b_n}{2}\right) e^{-in} 
$$

Using Euler Formula

$$
    f(x) = \sum_{k=0}^{k=K}  \left( \frac{a_n - i b_n}{2}\right) \left(Cos(n) + i Sin (n)\right) + \sum_{k=0}^{k=K} \left( \frac{a_n + ib_n}{2}\right) \left(Cos(n) - i Sin (n)\right) 
$$

$$
f(x) =  \sum_{k=0}^{k=K}\left( \frac{a_n + a_n}{2}\right) Cos(n) +  \sum_{k=0}^{k=K} i \left( \frac{a_n - a_n}{2}\right) Sin(n) +  \sum_{k=0}^{k=K} i \left( \frac{-b_n + b_n}{2}\right) Cos(n) +  \sum_{k=0}^{k=K} \left( \frac{b_n + b_n}{2}\right) Sin(n)
$$

$$
f(x) =  \sum_{k=0}^{k=K}a_n Cos(n) + \sum_{k=0}^{k=K} b_n Sin(n)
$$

### Define Complex Coefficients

&emsp; &emsp; Note that the inner product invlovling complex numbers is defined as :

&emsp; &emsp; $
\langle \mathbf{u}, \mathbf{v} \rangle = \sum_{i=1}^n u_i \overline{v_i}
$


#### &emsp;&emsp;Calculate the complex coefficients using projection

&emsp; &emsp; $
\text{proj}_{\mathbf{v}} \mathbf{u} = \frac{\langle \mathbf{u}, \mathbf{v} \rangle}{\langle \mathbf{v}, \mathbf{v} \rangle} \mathbf{v}
$

&emsp; &emsp; where $ \langle \mathbf{u}, \mathbf{v} \rangle $ is the Hermitian inner product of $ \mathbf{u} $ and $ \mathbf{v} $, defined as:

&emsp; &emsp; $
\langle \mathbf{u}, \mathbf{v} \rangle = \sum_{i=1}^n u_i \overline{v_i}
$

$$
f(x) = \sum_{k=-k}^{k=K} C_k e^{in} 
$$|

##### &emsp; The inner product of two **real** functions \(a(x)\) and \(b(x)\) over the **real number interval** \([a, b]\) is given by:

&emsp;&emsp;&emsp;&emsp; $$ \langle a, b \rangle = \int_a^b a(x) b(x) \, dx $$

##### &emsp; The inner product of two **complex** functions \(a(x)\) and \(b(x)\) over the **complex number interval** \([a, b]\) is given by:
&emsp;&emsp;&emsp;&emsp; $$ \langle a, b \rangle = \int_a^b a(x) \overline{b(x)} \, dx $$

#### &emsp;&emsp;Projecting f(x) on to the base $e^{in}$ to find $C_k$
&emsp; &emsp; $$
\text{proj}_{\mathbf{e^{in}}} \mathbf{f(x)} = \frac{\langle \mathbf{f(x)}, \mathbf{e^{in}} \rangle}{\langle \mathbf{e^{in}}, \mathbf{e^{in}} \rangle} \mathbf{e^{in}}
$$

##### &emsp;&emsp;Inner product formula
&emsp; &emsp; $$
C_k = \frac{\langle \mathbf{f(x)}, \mathbf{e^{in}} \rangle}{\langle \mathbf{e^{in}}, \mathbf{e^{in}} \rangle}
= \frac{\sum_{x=-L}^{x=L} f(x) \overline{e^{in}}} {\sum_{x=-L}^{x=L} e^{in}\overline{e^{in}}}
= \frac{\sum_{x=-L}^{x=L} f(x) e^{-in}} {\sum_{x=-L}^{x=L} 1}
= \frac{\sum_{x=-L}^{x=L} f(x) e^{-in}} {2L}
$$

##### &emsp; &emsp; Integration formula
$$
C_k = \frac{\langle \mathbf{f(x)}, \mathbf{e^{in}} \rangle}{\langle \mathbf{e^{in}}, \mathbf{e^{in}} \rangle} = 
= \int_{-L}^L f(x) \overline{e^{in}} \, dx
$$

```python
def dft_map(X, Fs, shift=True):
    resolution = float(Fs) / len(X)
    if shift:
        n, Y = dft_shift(X)
    else:
        Y = X
        n = np.arange(0, len(Y))
    f = n * resolution
    return f, Y
```

```python
n = 64
L = 30
dx = L/n
x = np.arange(-L/2,L/2,dx,dtype = "complex_")
f = np.cos(x) * np.exp(-np.power(x,2)/25)
df = -(np.sin(x) * np.exp(-np.power(x,2)/25) + (2/25) * x *f)
```

```python
dfFD = np.zeros(len(df),dtype = "complex_")
for kappa in range(len(df)-1):
    dfFD[kappa] = (f[kappa+1]-f[kappa])/dx
dfFD[-1] = dfFD[-2]
```

```python
fhat = np.fft.fft(f)
kappa = (2*np.pi/L) *np.arange(0,n)
kappa = np.fft.ifftshift(kappa)
dfhat = kappa * fhat * (1j)
dfFFT = np.real(np.fft.ifft(dfhat))
```

```python
plt.plot(x,df.real,color="k")
plt.plot(x,dfFD.real,'--',color = "b")
plt.plot(x,dfFFT.real,"--",color="c")
plt.legend(["Real")
plt.show()
```
